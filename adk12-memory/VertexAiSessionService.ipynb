## **<font color="red">VertexAiSessionService</font>**
- **How it works:** Uses Google Cloud Vertex AI infrastructure via API calls for session management. It only works with `Vertex AI`.
- **Persistence:** Yes. Data is managed reliably and scalably via <u>Vertex AI Agent Engine</u>.
- **Requires:**
  - A Google Cloud project (`pip install vertexai`)
  - A Google Cloud storage bucket that can be configured by this step.
  - A Reasoning Engine resource name/ID.
  - If you do not have a Google Cloud project and you want to try the `VertexAiSessionService` see <u>Vertex AI Express Mode</u>.
- **Best for:** Scalable production applications deployed on Google Cloud, especially when integrating with other Vertex AI features.

```python
# Required: pip install google-adk[vertexai]
# Plus GCP setup and authenticaiton

from google.adk.sessions import VertexAiSessionService
from dotenv import load_dotenv, find_dotenv
import os

load_dotenv(find_dotenv())

PROJECT_ID = os.getenv("PROJECT_ID")
LOCATION = os.getenv("GOOGLE_CLOUD_LOCATION")

REASONING_ENGINE_APP_NAME = f"projects/{PROJECT_ID}/locations/{LOCATION}/publishers/google/models/gemini-2.5-flash"

session_service = VertexAiSessionService(project=PROJECT_ID, location=LOCATION)
session_service
```
