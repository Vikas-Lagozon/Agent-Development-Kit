# **<font color="red">Audio to Text</font>**
- Now, we can use the AI to create real-time, voice-driven agents. However, these systems need low latency, two-way communication, real-time information retrieval, and the ability to handle complex tasks.
- Build one using *Gemini* and the *Google Agent Development Kit* (ADK).
- First, we create an agent with a persona but no access to external tools. This is the simplest agent, relying only on its pre-trained knowledge.

### **<font color="blue">Single TTS Audio to Text</font>**

```python
import os
import asyncio
from google.adk.sessions import InMemorySessionService
from google.adk.artifacts import InMemoryArtifactService
from google.adk.runners import Runner
from google.adk.agents import LlmAgent
import google.genai.types as types
from config import config

# -----------------------------
# Setup API Key
# -----------------------------
os.environ["GOOGLE_API_KEY"] = config.GOOGLE_API_KEY.strip()

# -----------------------------
# Create LlmAgent
# -----------------------------
MODEL = "gemini-2.5-flash"

audio_agent = LlmAgent(
    model=MODEL,
    name="audio_transcriber",
    instruction="""
    You are a speech-to-text agent.
    Transcribe the provided audio accurately.
    If multiple speakers are present, label them as Speaker 1, Speaker 2, etc.
    """
)

# -----------------------------
# Setup In-Memory Services
# -----------------------------
APP_NAME = "audio_app"
USER_ID = "user_001"
SESSION_ID = "audio_session"

session_service = InMemorySessionService()
artifact_service = InMemoryArtifactService()

await session_service.create_session(
    app_name=APP_NAME,
    user_id=USER_ID,
    session_id=SESSION_ID
)

# -----------------------------
# Setup Runner
# -----------------------------
runner = Runner(
    agent=audio_agent,
    app_name=APP_NAME,
    session_service=session_service,
    artifact_service=artifact_service
)

# -----------------------------
# Chat function for Audio → Text
# -----------------------------
async def transcribe_audio(file_path):
    with open(file_path, "rb") as f:
        audio_bytes = f.read()

    content = types.Content(
        role="user",
        parts=[
            types.Part(
                inline_data=types.Blob(
                    mime_type="audio/wav",  # change to mp3 if needed
                    data=audio_bytes
                )
            )
        ]
    )

    events = runner.run_async(
        user_id=USER_ID,
        session_id=SESSION_ID,
        new_message=content
    )

    async for event in events:
        if event.is_final_response() and event.content and event.content.parts:
            print("Transcription:\n", event.content.parts[0].text)

# -----------------------------
# Run example
# -----------------------------
await transcribe_audio("Impact_of_AI.wav")
```

```python
import os
import asyncio
from google.adk.sessions import InMemorySessionService
from google.adk.artifacts import InMemoryArtifactService
from google.adk.runners import Runner
from google.adk.agents import LlmAgent
import google.genai.types as types
from config import config

# -----------------------------
# Setup API Key
# -----------------------------
os.environ["GOOGLE_API_KEY"] = config.GOOGLE_API_KEY.strip()

MODEL = "gemini-2.5-flash"

APP_NAME = "audio_to_discussion_app"
USER_ID = "user_001"
SESSION_ID = "audio_session"

# -----------------------------
# Single Smart Agent
# -----------------------------
discussion_extractor_agent = LlmAgent(
    model=MODEL,
    name="discussion_extractor",
    instruction="""
You are an advanced speech-to-text podcast extractor.

Your task:
Convert the provided multi-speaker audio into a clean discussion transcript.

IMPORTANT RULES:

1. Accurately transcribe the audio.
2. Detect speaker changes carefully.
3. Detect real speaker names if spoken.
4. If names are detected (e.g., Vikas, Rohit), preserve them.
5. If no names are detected, use consistent labels:
   Speaker 1, Speaker 2, Speaker 3, etc.
6. Maintain original conversational flow.
7. Do NOT summarize.
8. Do NOT rewrite meaning.
9. Do NOT shorten.
10. Do NOT add narration or commentary.

STRICT OUTPUT FORMAT:

Name: dialogue
Name: dialogue
Name: dialogue

Return ONLY the formatted discussion text.
"""
)

# -----------------------------
# Setup In-Memory Services
# -----------------------------
session_service = InMemorySessionService()
artifact_service = InMemoryArtifactService()

async def setup_session():
    await session_service.create_session(
        app_name=APP_NAME,
        user_id=USER_ID,
        session_id=SESSION_ID
    )

# -----------------------------
# Create Runner
# -----------------------------
runner = Runner(
    agent=discussion_extractor_agent,
    app_name=APP_NAME,
    session_service=session_service,
    artifact_service=artifact_service
)

# -----------------------------
# Process Audio → Discussion
# -----------------------------
async def extract_discussion_from_audio(file_path):

    # Read audio file
    with open(file_path, "rb") as f:
        audio_bytes = f.read()

    content = types.Content(
        role="user",
        parts=[
            types.Part(
                inline_data=types.Blob(
                    mime_type="audio/wav",  # change to audio/mpeg if mp3
                    data=audio_bytes
                )
            )
        ]
    )

    events = runner.run_async(
        user_id=USER_ID,
        session_id=SESSION_ID,
        new_message=content
    )

    async for event in events:
        if event.is_final_response() and event.content and event.content.parts:
            discussion_text = event.content.parts[0].text

            print("\n💬 Final Discussion Output:\n")
            print(discussion_text)

            # Save to file
            with open("final_discussion.txt", "w", encoding="utf-8") as f:
                f.write(discussion_text)

            return discussion_text


# -----------------------------
# Run
# -----------------------------
async def main():
    await setup_session()
    await extract_discussion_from_audio("AI with Quantum Power.wav")  # <-- your file here

await main()
```
